```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statistics

all_data = pd.read_csv("all_data.csv")
print(all_data.head())
columns = all_data.columns.tolist()
```

```python
year_list = all_data["Year"].tolist()[:16] #We only want the first 16 entries of this column because of the repition that occurs with each country
print(year_list)
```

```python
#created empty lists to add life expectency for each country to - 
chile_life_expectency = []
china_life_expectency = []
germany_life_expectency = []
mexico_life_expectency = []
usa_life_expectency = []
zimbabwe_life_expectency = []
```

```python
#this loop adds all life expectancies to the appropriate list for each country
for index, row in all_data.iterrows():
    if row['Country'] == "Chile":
        chile_life_expectency.append(row['Life expectancy at birth (years)'])
    elif row['Country'] == "China":
        china_life_expectency.append(row['Life expectancy at birth (years)'])
    elif row['Country'] == "Germany":
        germany_life_expectency.append(row['Life expectancy at birth (years)'])
    elif row['Country'] == "Mexico":
        mexico_life_expectency.append(row['Life expectancy at birth (years)'])
    elif row['Country'] == "United States of America":
        usa_life_expectency.append(row['Life expectancy at birth (years)'])
    elif row['Country'] == "Zimbabwe":
        zimbabwe_life_expectency.append(row['Life expectancy at birth (years)'])

print(chile_life_expectency)
print(china_life_expectency)
```

```python
#added all the lits of life expectencies to one list in order to create a loop to create a line plot for each
life_expectencies = [
    [chile_life_expectency, "Chile"], [china_life_expectency, "China"], [germany_life_expectency, "Germany"], [mexico_life_expectency, "Mexico"], [usa_life_expectency, "USA"], [zimbabwe_life_expectency, "Zimbabwe"]
]

for list in life_expectencies:
    plt.plot(list[0], label=list[1])
    plt.xticks(rotation = 30, fontsize = 9)
    plt.xlabel("Years After 2000 ")
    plt.ylabel("Life Expectecy")
    plt.title("Different Countries' Life Expectency Over 15 Years")
    plt.ylim(40, 85)
    plt.legend()
plt.show()
```

```python
chile_gdp = []
china_gdp = []
germany_gdp = []
mexico_gdp = []
usa_gdp = []
zimbabwe_gdp = []
```

```python
#loop adds all entries of GDP to a list for each country
for index, row in all_data.iterrows():
    if row['Country'] == "Chile":
        chile_gdp.append(row['GDP'])
    elif row['Country'] == "China":
        china_gdp.append(row['GDP'])
    elif row['Country'] == "Germany":
        germany_gdp.append(row['GDP'])
    elif row['Country'] == "Mexico":
        mexico_gdp.append(row['GDP'])
    elif row['Country'] == "United States of America":
        usa_gdp.append(row['GDP'])
    elif row['Country'] == "Zimbabwe":
        zimbabwe_gdp.append(row['GDP'])
```

```python
gdp_list = [
    [chile_gdp, "Chile"], [china_gdp, "China"], [germany_gdp, "Germany"], [mexico_gdp, "Mexico"], [usa_gdp, "USA"], [zimbabwe_gdp, "Zimbabwe"]
]

for list in gdp_list: 
    plt.plot(list[0], label=list[1])
    plt.xticks(rotation = 30, fontsize = 9)
    plt.xlabel("Years After 2000")
    plt.ylabel("GDP")
    plt.title("Different Countries GDP Over 15 Years")
    plt.ylim(all_data["GDP"].min(), all_data["GDP"].max())
    plt.legend(loc = "upper left")
plt.show() #this line is not in the loop so that all the lines are on one single plot
    
```

```python
for list in life_expectencies:
    print("The average life expectency in " + list[1] + " from 2000 to 2015 is " + str(round(statistics.mean(list[0]) , 2)) + " years")
```

```python
gdp_and_life = [
    [chile_life_expectency, chile_gdp, "Chile"], [china_life_expectency, china_gdp, "China"], [germany_life_expectency, germany_gdp, "Germany"], [mexico_life_expectency, mexico_gdp, "Mexico"], [usa_life_expectency, usa_gdp, "USA"], [zimbabwe_life_expectency, zimbabwe_gdp, "Zimbabwe"]
]
for list in gdp_and_life: #Here I have each country on its own plot. This helps to see each individual countries' change instead of comparing. 
    plt.scatter(x = list[0], y = list[1])
    plt.title(list[2] + " GDP vs. Life Expectency")
    plt.xlabel("Life Expectency")
    plt.ylabel("GDP")
    plt.show()
```

```python
countries = ["Chile", "China", "Germany", "Mexico", "USA", "Zimbabwe"]
for list in gdp_and_life: #Here is the same plots but put on a single graph for easier comparison
    sns.scatterplot(x = list[0], y = list[1])
    plt.title("GDP vs. Life Expectency")
    plt.xlabel("Life Expectency")
    plt.ylabel("GDP")
    plt.legend(countries)
plt.show()
```
